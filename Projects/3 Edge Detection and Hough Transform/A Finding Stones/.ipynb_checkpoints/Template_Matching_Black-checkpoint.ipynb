```python
from PIL import Image, ImageStat          # Python Imaging Library
import numpy as np              # Numerical Python
import matplotlib.pyplot as plt # Python plotting
```

```python
# Helper Functions

# Generate Gray Scale Numpy Array from Image
def create_grayscale( my_image ):
    grayscale = np.array(Image.open( my_image ).convert('L'))
    return grayscale

# show grayscale image
def show_grayscale( my_image ):
    plt.figure(figsize = (10,10))
    plt.imshow(my_image, cmap = plt.get_cmap(name = 'gray'))
    plt.show()
```

```python
# Generate Mean of Image

#a = ImageStat.Stat(img)
```

```python
# Convert all images to grayscale
token_grayscale = create_grayscale( '../token_black_1.jpg')

go_grayscale = create_grayscale( '../go1.jpg')

show_grayscale(token_grayscale)
show_grayscale(go_grayscale)

a = ImageStat.Stat(token_grayscale)
print(a.mean)
```
